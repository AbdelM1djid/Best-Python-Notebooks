```python
data = pd.get_dummies(data,columns=["sex","fbs","restecg","exang","ca","thal","cp"])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
columns_toscale = ["age","trestbps","chol","thalach","oldpeak"]
data[columns_toscale] = scaler.fit_transform(data[columns_toscale])
```

```python
y = data.target
data.drop(["target"],inplace=True,axis=1)
X = data
```

```python
X.describe()
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test  = train_test_split(X,y,random_state=3,test_size=0.33)
```

```python
from sklearn.neighbors import KNeighborsClassifier
Knn_scores = []
for k in range (1,21) : 
    knc = KNeighborsClassifier(n_neighbors=k)
    knc.fit(X_train,y_train)
    score = knc.score(X_test,y_test)
    Knn_scores.append(score)
```

```python
plt.subplots(figsize=(15,15))
#next code to mark down values text in the plot
for i in range(1,21):
    plt.text(i, Knn_scores[i-1], (i, Knn_scores[i-1]))
sns.lineplot(x=[ i for i in range(1,21)],y=Knn_scores,markers=True,dashes=False)
plt.title('K Neighbors Classifier scores for different K values')
```

```python
from sklearn.svm import SVC
svc_scores = []
#basically making diffrent hyperplanes and testing each svc score for each one 
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
for i in range(len(kernels)):
    svc_classifier = SVC(kernel = kernels[i])
    svc_classifier.fit(X_train, y_train)
    svc_scores.append(svc_classifier.score(X_test, y_test))
```

```python
colors = rainbow(np.linspace(1,0,len(kernels)))
plt.bar(kernels,svc_scores,color=colors)
for i in range(len(kernels)) : 
    plt.text(i,svc_scores[i],svc_scores[i])
```
